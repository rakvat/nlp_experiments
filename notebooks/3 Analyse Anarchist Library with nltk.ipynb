```python
import glob

list_of_files = glob.glob('/home/rakvat/mydev/nlp_experiments/scrape_anarchist_library/en/*.txt') 
documents = {}
for filename in list_of_files:
    content = []
    f = open(filename, 'r')
    for line in f:
        content.append(line)
    f.close()
    documents[filename.split('/')[-1]] = content
```

## Preprocessing


```python
for key, value in documents.items():
    document = {}
    first_line_without_meta = 0
    for i, line in enumerate(value[:15]):
        if line.startswith('#author'):
            document['author'] = line[8:-1]
        if line.startswith('#title'):
            document['title'] = line[7:-1]
        if first_line_without_meta == 0 and line[0] != '#':
            first_line_without_meta = i
    document['text'] = ''.join(value[first_line_without_meta:])
    
    documents[key] = document
```

## nltk

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

default_stopwords = set(nltk.corpus.stopwords.words('english'))
default_stopwords = default_stopwords.union(['em', '/em', '/quote', '*****', '****', '***', "''", '``', 'br', '/strong'])
```

```python
from nltk.stem.snowball import SnowballStemmer

for key, document in documents.items():
    words = nltk.word_tokenize(document['text'])
    words = [word for word in words if len(word) > 1]
    words = [word.lower() for word in words]

    # not happy with the results of this
    #stemmer = nltk.stem.snowball.SnowballStemmer('english')
    #words = [stemmer.stem(word) for word in words]
    wnl = nltk.WordNetLemmatizer()
    words = [wnl.lemmatize(word) for word in words]

    words = [word for word in words if word not in default_stopwords]
    fdist = nltk.FreqDist(words)
    print(key, ' ### ', fdist.most_common(10), '\n')
```

```python
from nltk.text import Text  

word = 'simulation'
        
for key, value in documents.items():
    tokens = nltk.word_tokenize(value["text"])
    text_list = Text(tokens)
    ci = nltk.ConcordanceIndex(text_list)
    if ci.offsets(word):
        print(key, ":")
        ci.print_concordance(word)
        print()
```
