```python
import glob

list_of_files = glob.glob('/home/rakvat/WebDev/nanoc/transform/content/**/*.md') 
documents = {}
for filename in list_of_files:
    content = []
    f = open(filename, 'r')
    for line in f:
        content.append(line)
    f.close()
    documents[filename.split('/')[-1]] = content
```

```python
import nltk
```

```python
nltk.word_tokenize("hello world")
```

## Preprocessing





```python
for key, value in documents.items():
    documents[key] = ''.join(value)
```

## nltk

```python
for key, value in documents.items():
    print(nltk.word_tokenize(value))
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

default_stopwords = set(nltk.corpus.stopwords.words('german'))
default_stopwords_en = set(nltk.corpus.stopwords.words('english'))
default_stopwords = default_stopwords.union(default_stopwords_en)
default_stopwords = default_stopwords.union(['dass', "''", '--', '/a', '__blank', 'target=', 'href=','id', 'beim', 'name=','class=','anchor', 'eigene','eigenen'])
```

```python
for key, value in documents.items():
    words = nltk.word_tokenize(value)
    words = [word for word in words if len(word) > 1]
    words = [word.lower() for word in words]

    # no good results with this
    # stemmer = nltk.stem.snowball.SnowballStemmer('german')
    # words = [stemmer.stem(word) for word in words]

    words = [word for word in words if word not in default_stopwords]
    fdist = nltk.FreqDist(words)
    print(key, ' ### ', fdist.most_common(10))
```

```python
from nltk.text import Text  

for key, value in documents.items():
    tokens = nltk.word_tokenize(value)
    text_list = Text(tokens)
    concordance = text_list.concordance("Anarchie")
    print(key, ' ###\n', concordance)
```
